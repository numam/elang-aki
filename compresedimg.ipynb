```python
import os
from PIL import Image
```

```python
MAX_SIZE_MB = 1
MAX_SIZE_BYTES = MAX_SIZE_MB * 1024 * 1024

INPUT_DIR = "src/asset"
OUTPUT_DIR = "src/assetresize"

os.makedirs(OUTPUT_DIR, exist_ok=True)
```

```python
def resize_until_under_1mb(img, output_path, format):
    scale = 1.0

    while True:
        width = int(img.width * scale)
        height = int(img.height * scale)

        resized = img.resize((width, height), Image.LANCZOS)

        if format == "JPEG":
            resized.save(output_path, format="JPEG", quality=95, optimize=True)
        else:  # PNG
            resized.save(output_path, format="PNG", optimize=True)

        if os.path.getsize(output_path) <= MAX_SIZE_BYTES:
            break

        scale *= 0.9  # perkecil bertahap 10%
```

```python
for filename in os.listdir(INPUT_DIR):
    if not filename.lower().endswith((".jpg", ".jpeg", ".png")):
        continue

    input_path = os.path.join(INPUT_DIR, filename)
    output_path = os.path.join(OUTPUT_DIR, filename)

    with Image.open(input_path) as img:
        format = img.format  # JPEG / PNG
        resize_until_under_1mb(img, output_path, format)

    print(f"✅ {filename} selesai")
```
